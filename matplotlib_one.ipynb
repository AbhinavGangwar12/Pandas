DRAWING A CHART IS NOT DIFFICULT 
THE DIFFICULT PART IS TO CHOOSE WHICH TYPE OF CHART TO TAKE TO REPRESENT THE DATA
THERE ARE MANY TYPES OF CHARTS AVAILABLE WHICH WE CAN USE TO REPRESENT DIFFERENT TYPE OF DATA 

WHY DO WE NEED PLOTS ?
THE MAIN TWO REASON WHY WE NEED PLOTS ARE
1. EXPLORATORY : TO EXPLORE MY DATAPOINTS AND TO OBSERVE DIFFERENT TRENDS AND PATTERNS BETWEEN THEM
2. EXPLAINATORY : TO HAVE A VISUAL REPRESENTATION OF DATA AND CAN BE EXPLAINED TO ANYONE THAT HOW IS MY DATA BEHAVING AND WHICH ATTRIBUTES DO THEY POSSES

DATA VISUALISATION BASICALLY CONTAINS TWO ASPECTS
1. ART : SELECTING RIGHT SCALE,AXIS,TIDES - CREATING VALID GRAPHS etc.
2. SCIENCE : WHICH KIND OF GRAPH WILL BE BEST SUITED FOR MY DATA

NOMENCLATURE-->
IN DATA VISUALISATION WE HAVE DIFFERENT NOMECLATURE FOR DIFFERENT THINGS
-- THE COLUMN HEADERS ARE CALLED AS FEATURES/VARIABLES
-- INDIVIDUAL ROWS ARE CALLED SAMPLES/RECORDS/DATAPOINTS 
[ WE WILL WORK ON THEM MATHEMATICALLY TOO ]

DATA CAN BE BROADLY CLASSIFIED INTO TWO CATEGORIES 
1. CATEGORICAL DATA < N >
2. CONTINUOUS DATA < C >

    1. CATEGORICAL DATA FURTHER IS DIVIDED INTO TWO TYPES
        a. NOMINAL DATA -- WHERE THE VALUES DONOT HAVE ANY RELATIONSHIP BETWEEN THEM
        b. ORDINAL DATA -- WE HAVE CERTAIN RELATION BETWEEN THE DATASETS eg. HIGH,MEDIUM,LOW
    2. CONTINUOUS DATA ITSELF FURTHER IS DIVIDED INTO TWO CATEGORIES 
        a. INTERVAL -- THE DATA HAVING DIRECTION AND MAGNITUDE ( VECTORS ) eg. 0(deg)C . IN THIS TYPE OF DATA ZERO DOESNOT MEAN THAT THE VALUE IS ZERO AS IN THE GIVEN EXAMPLE IT IS VERY CLEAR
        b. RATIO whereas in RATIO zero means absolute zero means nothing-->0

ON THE BASIS OF NUMBER OF VARIABLES WE HAVE, THERE ARE 
1. UNIVARIATE -- WHEN WE ONLY HAVE ONE KIND OF VARIABLE, IT CAN BE 
    C
    N
2. BIVARIATE -- WHEN WE HAVE TWO KIND OF VARIABLES, THEN IT CAN BE EITHER
    CC
    NN
    NC
3. MULTIVARIATE -- WHEN WE HAVE N TYPE OF VARIABLES, THEN .. 
    NNNN...
    CCCC...
    CNCN...
    AND SO ON. SINCE IT BECOME VERY MESSI TO PLOT A CHART OF MORE THAN 3 VARIABLES SO WE KEEP IT UPTO THREE VARIABLES 
    THEREFORE WHEN WE TALK ABOUT THREE VARIABLE THEN WE CAN HAVE 
    NNN
    CCC
    NCN
    CNC
    NCC
    CNN

```python
df = pd.read_csv("final_vg.csv")
# df.loc["1942"]
# df.set_index('Name').loc["Zumba Fitness: World Party"]
# df.reset_index()
df.head(5)
```

UNIVARIATE ANALYSIS

```python
platforms = df["Platform"].value_counts()
platforms
```

```python
x = platforms.index
y = platforms
mp.bar(x,y)
```

```python
mp.figure(figsize=(15,10))
mp.bar(x,y)
```

```python
# IF YOUR WANT TO PERFORM CHANGES IN THE TICKS ON X-AXIS JUST USE THIS FUNCTION
# mp.xticks(rotation=-90)
mp.bar(x,y)
mp.xticks(rotation=90)#will show additional information regarding the xticks
mp.show()#will show the final plot after doing all the customisations above it and will avoid uneccessary information
```

```python
mp.figure(figsize=(10,8))#length,height
mp.bar(x,y,width=0.7,color="Red")#changing color of bars and width of the bar
mp.title("Platforms for Games")
mp.xlabel("Platforms")
mp.ylabel("Count")
mp.xticks(rotation=90,fontsize=10)
mp.show()
```

```python
sns.countplot(data=df,order = df["Platform"].value_counts().index,x = "Platform",color="Red")
sns.x
```

IN SEABORN, BARGRAPH --> COUNTPLOT
IT AUTOMATICALLY COUNTS THE UNIQUE VALUES IN THE GIVEN COLUMN AND MAKE A COUNT BAR GRAPH OUT OF IT
COUNTPLOT AUTOMATICALLY COUNTS THE NUMBER OF UNIQUE VALUES WHEREAS IN MATPLOTLIB WE HAVE TO PROVIDE THE COUNTS
COUNTPLOT AUTOMATICALLY DOES MANY A TASK LIKE IT CREATES XLABEL AND YLABEL ON IT'S ON
THEREFORE SEABORN IS SMARTER

NUMERICAL / CONTINUOUS UNIVARIATE DATA
HISTOGRAMS ARE CAN BE USED DISPLAY CONTINUOUS DATA 

```python
mp.hist(df["Year"])#It is automatically counting the frequencies within ranges and it makes ranges automatically from the passed columns
# Similar to pd.cut() it makes bins and fill those bins

mp.show()
```

Here we have continous bars which show that between that particular period we have certain number of frequencies


```python
mp.hist(df["Year"],bins=5)
mp.show()
```

HISTOGRAM RETURN THREE THINGS COUNTS,BINS AND OTHER VALUES
BINS - CONTAINS THE BINS LIKE WHAT ACTUALLY THE BINS ARE
COUNTS - CONTAINS THE VALUES THE ARE IN THERE CORRESPONDING BINS

```python
sns.histplot(df["Year"],bins=4)
```

```python
sns.histplot(data=df,y="Year",bins=5,color="red")
mp.xticks(rotation = 90)
mp.show()
```

```python
sns.kdeplot(df["Year"],color="red")#THis is a type of histogram but it's a kinda smooth. This basically helps in estimating probability
mp.show()
```
