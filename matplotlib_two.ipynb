```python
df = pd.read_csv("final_vg.csv")
df
```

```python
data = df.head(50)
plt.figure(figsize=(7,5))
sns.boxplot(data = data,y="JP_Sales",color="Yellow")
```

```python
plt.figure(figsize=(7,5))
sns.boxplot(data = data,y="Global_Sales",color="red")#there are no outliers in this data
plt.show()
```

```python
ih = df.loc[df["Name"] == "Ice Hockey"]
bb = df.loc[df["Name"] == "Baseball"]
nfs = df.loc[df["Name"] == "Need for Speed: Most Wanted"]

sns.lineplot(data=ih,y="NA_Sales",x="Year",color="purple",label="Ice Hockey")
sns.lineplot(data=bb,y="NA_Sales",x="Year",color="yellow",label="Ice Hockey")
# sns.lineplot(data=nfs,y="NA_Sales",x="Year",color="red")
plt.title("North American Games sales")
plt.show()
```

```python
sns.lineplot(data=ih,y="NA_Sales",x="Year",color="purple",label="Ice Hockey")
sns.lineplot(data=bb,y="NA_Sales",x="Year",color="yellow",label="Ice Hockey")
sns.lineplot(data=nfs,y="NA_Sales",x="Year",color="red",label="NFS")
plt.xlim(left=2005,right=2014)#We are limiting the values of x-axis 
plt.ylim(bottom=0,top=4)#here we are limiting the values of y-axis
plt.legend(loc=(-0.5,0),ncol=2)#considers whole graph as one unit
plt.grid()
plt.title("North American Games sales")
plt.show()
```

```python
top_pub = df["Publisher"].value_counts().index[:3]
top_plt = df["Platform"].value_counts().index[:3]
top_gen = df["Genre"].value_counts().index[:3]

top_data = df.loc[(df["Platform"].isin(top_plt)) & (df["Publisher"].isin(top_pub)) & (df["Genre"].isin(top_gen))]
top_data
```

```python
sns.scatterplot(data=top_data,x="NA_Sales",y="Global_Sales")
plt.show()
```

```python
# Now you wan to check NA SAles and Global Sales with respect to genre of the games like which genre game 
sns.scatterplot(data=top_data,x="NA_Sales",y="Global_Sales",hue="Genre")
```

```python
sns.scatterplot(data=top_data,x="JP_Sales",y="Global_Sales")
```

```python
sns.scatterplot(data=top_data,x="Other_Sales",y="Global_Sales")
```

```python
sns.lineplot(data=top_data,x="Year",y="Global_Sales",hue="Genre")
```

CATEGORICAL - CONTINUOUS

```python
sns.countplot(data=top_data,x="Publisher",hue="Genre")#this is dodged bar chart
# sns.countplot(data=top_data,x="Publisher")
plt.show()
```

```python
sns.boxplot(data=top_data,y="Global_Sales",x="Publisher",color="orange")
# ONE IS CATEGORICAL AND OTHER IS CONTINUOUS
```

```python
sns.violinplot(data=top_data,y="Global_Sales",hue="Genre")

#this graph shows the distribution of data among the range of box plot
# there is box at the centre and the distribution of data like how much data is of which values and where it is lying, this is displayed by that the body
```

```python
sns.barplot(data=top_data,x="Platform",y="Global_Sales",estimator=np.mean)
plt.show()
sns.violinplot(data=top_data,y="EU_Sales",hue="Platform")
plt.show()
```

```python
# CATEGORICAL - CATEGORICAL - CONTINOUS
sns.violinplot(data=top_data,x="Publisher",y="Global_Sales",hue="Genre")
plt.show()
```

```python
# CONTINUOUS - CONTINUOUS - CONTINOUS
sns.scatterplot(data=top_data,x="NA_Sales",y = "EU_Sales", hue="Rank")
```

```python
# We can have different types of bubbles    
sns.scatterplot(data=top_data,x="NA_Sales",y = "EU_Sales", size="Rank")
```

```python
plt.figure(figsize=(6,5))
sns.scatterplot(data=top_data,x="NA_Sales",y = "EU_Sales", size="Rank",hue="Platform")
plt.legend(loc=(-0.35,0))
plt.show()
```

```python
plt.figure(figsize=(6,5))
sns.scatterplot(data=top_data,x="NA_Sales",y = "EU_Sales", size="Rank",hue="Platform",sizes=(1,200))
plt.legend(loc=(-0.35,0))
plt.show()
```

```python
# This is how we create multiple charts in one chart
plt.figure(figsize=(12,10))
plt.subplot(2,3,1)
sns.scatterplot(data=top_data,x="NA_Sales",y = "EU_Sales", size="Rank",hue="Platform")
plt.legend(loc=(-0.35,0))
plt.subplot(2,3,3)
sns.violinplot(data=top_data,x="Publisher",y="Global_Sales",hue="Genre")
plt.subplot(2,3,5)
sns.scatterplot(data=top_data,x="NA_Sales",y = "EU_Sales", hue="Rank")
plt.show()
```
